```python
#load relevant modules, set plotting to inline
import pyart
from matplotlib import pyplot as plt
import numpy as np
import scipy.ndimage as ndi
import wradlib
print pyart.__version__
%matplotlib inline
```

```python
sample_file = '/data/sample_sapr_data/235545.mdv'
radar = pyart.io.read(sample_file)
```

```python
myd = pyart.graph.RadarMapDisplay(radar)
fig = plt.figure(figsize = [18,10])
myd.plot_ppi_map('reflectivity', vmin = -8, vmax = 64, resolution = 'l')
```

```python
testim = wradlib.clutter.filter_gabella(radar.fields['reflectivity']['data'])
```

```python
clutter_fld = radar.fields['reflectivity'].copy()
clutter_fld['data'] = testim
clutter_fld['standard_name'] = 'clutter mask'
clutter_fld['units'] = 'unitless'
radar.add_field('clutter_gab', clutter_fld)
```

```python
myd = pyart.graph.RadarMapDisplay(radar)
fig = plt.figure(figsize = [18,10])
myd.plot_ppi_map('clutter_gab', resolution = 'l', vmin = 0, vmax = 1)
```
