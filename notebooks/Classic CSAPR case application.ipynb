```python
#load relevant modules, set plotting to inline
import pyart
from matplotlib import pyplot as plt
import numpy as np

print pyart.__version__
%matplotlib inline
```

```python
sample_file = '/data/sample_sapr_data/235545.mdv'
radar = pyart.io.read(sample_file)
```

```python
myd = pyart.graph.RadarMapDisplay(radar)
fig = plt.figure(figsize = [18,10])
myd.plot_ppi_map('reflectivity', vmin = -8, vmax = 64, resolution = 'l')
```

```python
print radar.fields.keys()
```

```python
for key in ['differential_phase', 'differential_reflectivity',  'velocity']:
    print key
    tex = pyart.retrieve.texture(radar.fields[key], 10)
    print tex['data'].shape
    radar.add_field('texture_of_' + key, tex)
    
```

```python
myd = pyart.graph.RadarMapDisplay(radar)
```

```python
fig = plt.figure(figsize = [18,10])
myd.plot_ppi_map('texture_of_differential_phase', vmin = 0, vmax = 100, resolution = 'l')
```
